## Soal 3 point 2 dan 3 

1. Tentukan variabel-variabel yang memiliki nilai korelasi paling positif, dan berikan penjelasan terkait nilai korelasi tersebut.
varibel yang memiliki nilai korelasi paling positif adalah variabel dengan warna biru pekat atau mendekati hitam. seperti parkinson dengan alzheimer, Drowning dengan meningitis, dan lainnya. lalu apa maksud dari korelasi positif ini Jika koefisien korelasi antara dua variabel mendekati 1, ini menunjukkan adanya hubungan linier positif yang kuat antara kedua variabel tersebut. Artinya, ketika salah satu variabel meningkat, kemungkinan besar variabel lainnya juga akan meningkat secara linier, dan sebaliknya, ketika salah satu variabel menurun, variabel lainnya kemungkinan besar akan menurun secara linier.
2. Tentukan variabel-variabel yang memiliki nilai korelasi paling negatif, dan berikan penjelasan terkait nilai korelasi tersebut.
varibel yang memiliki nilai korelasi paling negatif adalah variabel dengan warna cream atau mendekati putih. seperti parkinson dengan acute hepatitis, HIV dengan acute hepatitis, dan lainnya. lalu jika koefisien korelasi antara dua variabel mendekati -1, ini menunjukkan adanya hubungan linier negatif yang kuat antara kedua variabel tersebut. Artinya, ketika salah satu variabel meningkat, kemungkinan besar variabel lainnya akan menurun secara linier, dan sebaliknya, ketika salah satu variabel menurun, variabel lainnya kemungkinan besar akan meningkat secara linier.


```python
fig = plt.figure() # create figure

ax0 = fig.add_subplot(1, 2, 1) # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(1, 2, 2) # add subplot 2 (1 row, 2 columns, second plot). See tip below**
x = df_indo_2['Parkinson']
y = df_indo_2['Acute hepatitis']

x2 = df_indo_2['Drowning']
y2 = df_indo_2['Meningitis']

ax0.scatter(x, y)
ax0.set_title('Korelasi Negatif')
ax0.set_xlabel('Parkinson')
ax0.set_ylabel('Acute hepatitis')

ax1.scatter(x2, y2)
ax1.set_title('Korelasi Positif')
ax1.set_xlabel('Drowning')
ax1.set_ylabel('Meningitis')
```

disini terlihat bagaimana perbedaan antara korelasi negatif dan korelasi positif yang memiliki perbedaan arah kelinearan

```python
df_use = df[['Area','Year','Value']]
```

```python
# Filter data untuk negara China, India, Indonesia, dan Thailand
df_China = df_use[df_use['Area']=='China'].groupby(['Area','Year'], as_index = False).sum()
df_Indo = df_use[df_use['Area']=='Indonesia'].groupby(['Area','Year'], as_index = False).sum()
df_india = df_use[df_use['Area']=='India'].groupby(['Area','Year'], as_index = False).sum()
df_Thai = df_use[df_use['Area']=='Thailand'].groupby(['Area','Year'], as_index = False).sum()
```

```python
df_merged = pd.merge(df_China, df_Indo, on='Year')
df_merged
```

```python
df_merged = df_merged.drop('Area_x', axis=1)
```

```python
df_merged 
```

```python
df_merged = df_merged.drop('Area_y',axis = 1)
```

```python
df_merged
```

```python
df_merged.rename(columns = {'Value_x':'China'}, inplace = True)
df_merged.rename(columns = {'Value_y':'Indonesia'}, inplace = True)
```

```python
df_merged
```

```python
df_merged = pd.merge(df_merged, df_india, on='Year')
df_merged
```

```python
df_merged = pd.merge(df_merged, df_Thai, on='Year')
df_merged
```

```python
df_merged.rename(columns = {'Value_x':'India'}, inplace = True)
df_merged.rename(columns = {'Value_y':'Thai'}, inplace = True)
df_merged = df_merged.drop('Area_y',axis = 1)
df_merged = df_merged.drop('Area_x',axis = 1)
```

```python
df_merged
```

## soal 4 point 1 

```python
fig, ax = plt.subplots(figsize=(20, 10))  # Mengatur ukuran gambar
x = df_merged['Year']
bar_width = 0.45

# Menggambar bar plot untuk Data_China
rects1 = ax.bar(df_merged['Year'] - 0.4, df_merged['China'], width=0.25, label='China')

# Menggambar bar plot untuk Indonesia
rects2 = ax.bar(df_merged['Year'] -  0.25, df_merged['Indonesia'], width=0.25, label='Indonesia')

# Menggambar bar plot untuk India
rects3 = ax.bar(df_merged['Year'], df_merged['India'], width=0.25, label='India')

# Menggambar bar plot untuk Thailand
rects4 = ax.bar(df_merged['Year'] + 0.25, df_merged['Thai'], width=0.25, label='Thailand')

# Menambahkan label pada sumbu x dan y
ax.set_xlabel('Year', fontsize=12)  # Mengatur ukuran font sumbu x
ax.set_ylabel('Production Value in thousand of tons', fontsize=12)  # Mengatur ukuran font sumbu y
ax.set_title('Production Value by Country', fontsize=14)  # Mengatur ukuran font judul
ax.legend(bbox_to_anchor=(1, 1), loc='upper left')

# Mengatur posisi dan label pada tick pada sumbu x
ax.set_xticks(x + 0.025*bar_width) # Mengatur posisi tick pada tengah-tengah grup bar
ax.set_xticklabels(df_merged['Year'], rotation = 'vertical') # Mengatur label tick dengan nilai dari kolom 'Year'

# Menampilkan plot
plt.show()
```

## soal 4 point 2

```python
correlation = df_merged[['China','Indonesia','India','Thai']].corr()

mask = np.triu(np.ones_like(correlation))
sns.heatmap(correlation, annot=True,mask=mask, cmap='coolwarm')

# Menampilkan plot
plt.title('Correlation Heatmap antara 4 Negara')
plt.show()
```

Apakah ada hubungan korelasi jumlah produksi antar negara-negara China, India, Indonesia, dan Thailand?
Ya tentu saja ada korelasi antara 4 negara tersebut. ini dapat dilihat dari hasil heatmap yang dimana masing-masing negara berkorelasi satu sama lain yang nilai korelasinya hampir sama besar yaitu mendekati nilai 1. ini menandakan bahwa ada hubungan yang erat antara jumlah produksi beras pada negara tersebut. korelasi yang mendekati arah positif menandakan bahwa adanya hubungan produksi beras antar 4 negara tersebut dimana ketika salah satu negara meningkat produksinya, kemungkinan besar negara lainnya juga akan meningkat produksinya secara linier, dan sebaliknya.
